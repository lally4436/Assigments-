```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('crime_data.csv')

# Display the first few rows of the dataframe
print(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
normalize=MinMaxScaler()
new_df= normalize.fit_transform(df.iloc[:,1:].to_numpy())
df_scaled = pd.DataFrame(new_df,columns=df.columns[1:])
df_scaled
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(18, 8))
dend = shc.dendrogram(shc.linkage(df_scaled, method='average'))
```

```python
#create Cluster
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters= 5,
                             affinity='euclidean',
                             linkage='average')
```

```python
#save cluster for chart
y_hc = hc.fit_predict(df_scaled)
y_hc
```

```python
clusters= pd.DataFrame(y_hc,columns=['clusters'])
clusters
```

```python
df_scaled['h_clusterid']=clusters
df_scaled
```

```python
df_scaled.sort_values("h_clusterid")
```

```python
df['h_clusterid']=clusters
df.sort_values("h_clusterid")
```

```python
from sklearn.cluster import KMeans

# Determine the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# Output:
# - We used KMeans clustering to segment the data into clusters.
# - We determined the optimal number of clusters using the elbow method.
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors

# Determine the optimal epsilon value using the k-distance graph
nn = NearestNeighbors(n_neighbors=2)
nbrs = nn.fit(df_scaled)
distances, indices = nbrs.kneighbors(df_scaled)
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.figure(figsize=(10, 6))
plt.plot(distances)
plt.title('k-distance graph')
plt.xlabel('Data Points')
plt.ylabel('Epsilon')
plt.show()

# Output:
# - We used DBSCAN clustering to segment the data into clusters.
# - We determined the optimal epsilon value based on the k-distance graph.
```
