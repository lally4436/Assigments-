```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

# Load the dataset
book_data = pd.read_csv('/content/book.csv')
my_movies_data = pd.read_csv('/content/my_movies.csv')

```

```python
# Convert the datasets into transaction format
book_transactions = [row.dropna().tolist() for index, row in book_data.iterrows()]
my_movies_transactions = [row.dropna().tolist() for index, row in my_movies_data.iterrows()]
```

```python
print(my_movies_transactions)
```

```python
# Convert all elements in my_movies_transactions to strings
my_movies_transactions = [list(map(str, transaction)) for transaction in my_movies_transactions]

# Alternatively, convert all elements in book_transactions to integers
book_transactions = [list(map(int, transaction)) for transaction in book_transactions]
```

```python
# Apply one-hot encoding
te = TransactionEncoder()
book_encoded = te.fit(book_transactions).transform(book_transactions)
my_movies_encoded = te.fit(my_movies_transactions).transform(my_movies_transactions)
```

```python
# Check the shape of book_encoded
print("Shape of book_encoded:", book_encoded.shape)

# Check the number of unique items obtained from TransactionEncoder
print("Number of unique items:", len(te.columns_))
```

```python
# Convert the encoded data back to dataframe
book_df = pd.DataFrame(book_encoded)
my_movies_df = pd.DataFrame(my_movies_encoded)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

def generate_rules(data, min_support, min_confidence, min_length):
    # Find frequent itemsets using Apriori algorithm
    frequent_itemsets = apriori(data, min_support=min_support, use_colnames=True)

    # Generate association rules
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

    # Filter rules based on minimum length
    rules = rules[rules['antecedents'].apply(lambda x: len(x) >= min_length)]

    return rules

# Try different values of support, confidence, and minimum length
support_values = [0.1, 0.2, 0.3]
confidence_values = [0.5, 0.6, 0.7]
min_length_values = [1, 2]

for min_support in support_values:
    for min_confidence in confidence_values:
        for min_length in min_length_values:
            print(f"Min Support: {min_support}, Min Confidence: {min_confidence}, Min Length: {min_length}")
            print("Book Data:")
            book_rules = generate_rules(book_df, min_support, min_confidence, min_length)
            print(book_rules)
            print("My Movies Data:")
            my_movies_rules = generate_rules(my_movies_df, min_support, min_confidence, min_length)
            print(my_movies_rules)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization of association rules
def plot_rules(rules, title):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='support', y='confidence', size='lift', data=rules)
    plt.title(title)
    plt.xlabel('Support')
    plt.ylabel('Confidence')
    plt.show()

# Plot rules for book data
plot_rules(book_rules, 'Association Rules for Book Data')

# Plot rules for my_movies data
plot_rules(my_movies_rules, 'Association Rules for My Movies Data')
```
