```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
cars_df = pd.read_csv('/content/Cars.csv')
```

```python
mpg_data = cars_df['MPG']
```

```python
plt.figure(figsize=(8, 6))
plt.hist(mpg_data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
shapiro_test_statistic, shapiro_p_value = stats.shapiro(mpg_data)
```

```python
print("Shapiro-Wilk Test:")
print("Test Statistic:", shapiro_test_statistic)
print("p-value:", shapiro_p_value)
```

```python
if shapiro_p_value > 0.05:
    print("Since p-value > 0.05, the MPG data likely follows a normal distribution.")
else:
    print("Since p-value <= 0.05, the MPG data does not appear to follow a normal distribution.")
```
