```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Load and Explore the Data
# Load the dataset
delivery_data = pd.read_csv("/content/delivery_time.csv")


```

```python
# Check the first few rows of the dataset
print(delivery_data.head())

# Check the summary information of the dataset
print(delivery_data.info())

# Check for missing values
print(delivery_data.isnull().sum())
```

```python

# Step 3: Exploratory Data Analysis (EDA)
# Explore the relationship between 'Sorting Time' and 'Delivery Time' using scatter plot
plt.scatter(delivery_data['Sorting Time'], delivery_data['Delivery Time'])
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.title('Scatter Plot of Sorting Time vs Delivery Time')
plt.show()
```

```python
# Calculate correlation between 'Sorting Time' and 'Delivery Time'
correlation = delivery_data['Delivery Time'].corr(delivery_data['Sorting Time'])
print("Correlation between Sorting Time and Delivery Time:", correlation)
```

```python
# Step 4: Model Building
# Build a simple linear regression model
from sklearn.linear_model import LinearRegression
```

```python

# Define independent and dependent variables
X = delivery_data[['Sorting Time']]
y = delivery_data['Delivery Time']
```

```python

# Create and fit the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Get the coefficients of the model
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)
```

```python
# Step 5: Model Evaluation (Optional)
# Predict delivery time using the model
predicted_delivery_time = model.predict(X)
```

```python
# Step 6: Visualization (Optional)
# Visualize the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, predicted_delivery_time, color='red')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.title('Linear Regression: Sorting Time vs Delivery Time')
plt.show()
```

```python
import numpy as np

sorting_time = np.array(delivery_data['Sorting Time']).reshape(-1, 1)

predicted_delivery_time = model.predict(sorting_time)

print("\nPredicted Delivery Time:")
print(predicted_delivery_time)
```

## **`Salary_Data Solution\**


```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Step 2: Load and Explore the Data
# Load the dataset
salary_data = pd.read_csv("/content/Salary_Data.csv")
```

```python
# Check the first few rows of the dataset
print(salary_data.head())

# Check the summary information of the dataset
print(salary_data.info())

# Check for missing values
print(salary_data.isnull().sum())
```

```python
# Step 3: Exploratory Data Analysis (EDA)
# Explore the relationship between 'YearsExperience' and 'Salary' using scatter plot
plt.scatter(salary_data['YearsExperience'], salary_data['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Years of Experience vs Salary')
plt.show()


```

```python
# Calculate correlation between 'YearsExperience' and 'Salary'
correlation = salary_data['Salary'].corr(salary_data['YearsExperience'])
print("Correlation between Years of Experience and Salary:", correlation)
```

```python
# Model Building
# Adding a constant term to the independent variable
X = sm.add_constant(salary_data['YearsExperience'])
y = salary_data['Salary']
```

```python
# Fit the regression model
model = sm.OLS(y, X).fit()
```

```python

# Model summary
print(model.summary())
```

```python
# Visualize the regression line
plt.scatter(salary_data['YearsExperience'], salary_data['Salary'], color='blue')
plt.plot(salary_data['YearsExperience'], model.predict(X), color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Regression Line of Salary vs Years of Experience')
plt.show()
```

```python
# Predicting salaries for new data points
new_years_experience = [5, 7, 10, 12]  # Example years of experience
new_X = sm.add_constant(new_years_experience)  # Adding constant term
predicted_salaries = model.predict(new_X)

# Display predicted salaries
for years_exp, salary in zip(new_years_experience, predicted_salaries):
    print(f"Years of Experience: {years_exp}, Predicted Salary: {salary}")
```
